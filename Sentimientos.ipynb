```python
import os
import pandas as pd
import numpy as np
import re
pd.set_option('display.max_columns', None)
```

```python
from pysentimiento import create_analyzer
```

```python
#An√°lisis de sentimientos
analyzer = create_analyzer(task="sentiment", lang="es")
```

```python
#An√°lisis de emociones
emoanalyzer = create_analyzer(task="emotion", lang="es")
```

```python
#Hate Speech 
hateanalyzer=create_analyzer(task="hate_speech", lang="es")
```

```python
#Cargar datos de Twitter
os.chdir('/Users/santiagoherreragarcia/OneDrive - Universidad de los Andes/Semestre 2022-1/Machine Learning/Trabajo Final/Data')
consolidated=pd.read_pickle('Consolidated_V0.pkl')
```

```python
consolidated.head()
```

```python
#A cada Tweet vamos a calcularle todas las probabilidades
def simple_pre(tweet):
    try:
        tweet=float(tweet)
        if np.isnan(tweet)==True:
            tweet=''
        else:
            tweet=str(tweet)
    except ValueError:
        if tweet is None:
            tweet=''
        else: 
            #Simple preprocessing
            tweet=re.sub('RT.*?:','',tweet).strip().lower()
            tweet=re.sub(r'@\w+', '', tweet)
            tweet=re.sub('\n+', ' ', tweet)
    return tweet
def analyze_tweets_all(tweet):
    #Sentimiento
    sent=analyzer.predict(tweet).probas
    #Emociones
    emos=emoanalyzer.predict(tweet).probas
    #Hate Speech
    hate=hateanalyzer.predict(tweet).probas
    #List
    res=[sent,emos,hate]
    return res
```

```python
descriptions=consolidated.loc[:,['Author ID', 'Author Description']].drop_duplicates()
```

```python
descriptions['pre_description']=descriptions['Author Description'].apply(lambda x: simple_pre(x))
```

```python
rows=[]
u=1
for i in descriptions['pre_description']:
    print((u/len(descriptions['pre_description']))*100)
    results=analyze_tweets_all(i)
    row={}
    if i !='':
        row['neutral']=results[0]['NEU']
        row['negative']=results[0]['NEG']
        row['positive']=results[0]['POS']
        row['joy']=results[1]['joy']
        row['sadness']=results[1]['sadness']
        row['anger']=results[1]['anger']
        row['surprise']=results[1]['surprise']
        row['disgust']=results[1]['disgust']
        row['fear']=results[1]['fear']
        row['hateful']=results[2]['hateful']
        row['targeted']=results[2]['targeted']
        row['aggressive']= results[2]['aggressive']
    else:
        row['neutral']=0
        row['negative']=0
        row['positive']=0
        row['joy']=0
        row['sadness']=0
        row['anger']=0
        row['surprise']=0
        row['disgust']=0
        row['fear']=0
        row['hateful']=0
        row['targeted']=0
        row['aggressive']=0       
    row=pd.DataFrame(pd.Series(row)).transpose()
    rows.append(row)
    u=u+1
all_results=pd.concat(rows)
```

```python
descriptions_analyzed=pd.concat([descriptions,all_results], axis=1)
```
