```python
marcas=pd.read_pickle('Y Botometer.pkl')
```

```python
marcas.head()
```

```python
from matplotlib import pyplot as plt
```

```python
df_plot=pd.DataFrame(marcas.bot.value_counts())
df_plot.index=['No bot (%)','Bot (%)']
df_plot.reset_index(inplace=True)
```

```python
fig=plt.figure(figsize =(8, 8))
plot1=plt.pie(df_plot['bot'],labels=df_plot['index'], autopct='%.2f', colors=['lightblue','#B7C3F3'])
plt.savefig('Marcas.png', dpi=200, bbox_inches='tight')
```
